```python
unsmile_labels = ['gender_bias', 'others_bias', 'none_bias', 'hate', 'offensive', 'none_hate']
```

### 2. Model load

```python
from transformers import BertForSequenceClassification, TrainingArguments, Trainer, AutoTokenizer
import torch
import numpy as np
```

```python
model_name = 'beomi/kcbert-base'
```

```python
# PLM을 이용한 토크나이저 객체 생성
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# multi label classification 학습을 위해선 label이 float 형태로 변형되어야 합니다.
# huggingface datasets 최신 버전에는 'map' 함수에 버그가 있어서 변형이 올바르게 되지 않습니다. 

def preprocess_function(examples):
    tokenized_examples = tokenizer(str(examples["comments"])) # 문장 칼럼 텍스트 토크나이징 진행
    tokenized_examples['labels'] = torch.tensor(examples["labels"], dtype=torch.float) # labels 칼럼 타입 변환

    return tokenized_examples
```

```python
# 참고: https://www.ohsuz.dev/22f4e8e7-64a3-4789-9dd2-171913883733
tokenized_dataset = ds.map(preprocess_function)
tokenized_dataset.set_format(type='torch', columns=['input_ids', 'labels', 'attention_mask', 'token_type_ids'])
# input_ids: 모델의 입력, attention_mask: padding 구분, token_type_ids: 문장 구분 짓는 id
```

```python
tokenized_dataset['train'][0]
```

```python
tokenized_dataset['train'][0]
```

```python
from transformers import DataCollatorWithPadding
# 토크나이징-패딩 진행 개체 생성
data_collator = DataCollatorWithPadding(tokenizer=tokenizer) 
```

```python
num_labels=len(unsmile_labels) # Label 갯수

# Pre train된 BERT 문장분류 모델 생성 (multi-label classification, 다중분류), label 개수 입력해주기
# id -> label, label -> id: 혐오 유형별 id 코드 부여
# BERT, Attention Mask? https://wikidocs.net/115055
model = BertForSequenceClassification.from_pretrained(
    model_name, 
    num_labels=num_labels, 
    problem_type="multi_label_classification"
)
model.config.id2label = {i: label for i, label in zip(range(num_labels), unsmile_labels)}
model.config.label2id = {label: i for i, label in zip(range(num_labels), unsmile_labels)}
```

```python
model.config.label2id
```

###3. Model Training

```python
from sklearn.metrics import label_ranking_average_precision_score
```

```python
# 모델 예측 (혐오지수 계산) 함수
def compute_metrics(x):
    return {
        'lrap': label_ranking_average_precision_score(x.label_ids, x.predictions),
    }
```

```python
batch_size = 64 # 64 batch는 colab pro에서 테스트되었습니다.
```

```python
# Train, Valid, Score Compute 진행하는 trainer 개체 생성
args = TrainingArguments(
    output_dir="model_output",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=1,
    save_strategy='epoch',
    load_best_model_at_end=True,
    metric_for_best_model='lrap',
    greater_is_better=True,
)

trainer = Trainer(
    model=model, 
    args=args, 
    train_dataset=tokenized_dataset["train"], 
    eval_dataset=tokenized_dataset["valid"], 
    compute_metrics=compute_metrics,
    tokenizer=tokenizer,
    data_collator=data_collator
)
```

```python
# 모델 훈련 진행 
trainer.train()
```

```python
trainer.save_model()
```

## 5. Model Test

```python
from transformers import TextClassificationPipeline

# 토크나이징, BERT multiple classification 모델 파이프라인 생성 -> 혐오지수 예측 진행
pipe = TextClassificationPipeline(
    model = model,
    tokenizer = tokenizer,
    device=0,
    return_all_scores=True,
    function_to_apply='sigmoid'
    )
```

```python
for result in pipe("이래서 여자는 게임을 하면 안된다")[0]:
    print(result)
```

## 5. Model evaluation

```python
# 모델 예측 결과 성능 평가 함수
def get_predicated_label(output_labels, min_score):
    labels = []
    for label in output_labels:
        if label['score'] > min_score:
            labels.append(1)
        else:
            labels.append(0)
    return labels
```

```python
import tqdm
from transformers.pipelines.base import KeyDataset

# 모델 성능 평가 진행
predicated_labels = []

for out in tqdm.tqdm(pipe(KeyDataset(ds['valid'], 'comments'))):
    predicated_labels.append(get_predicated_label(out, 0.5))
```

```python
from sklearn.metrics import classification_report

print(classification_report(ds['valid']['labels'], predicated_labels))
# 위 모델은 예측률 78%, 재현율 78%, F1 스코어 77%, 지지도 3985로 평가됨
```
